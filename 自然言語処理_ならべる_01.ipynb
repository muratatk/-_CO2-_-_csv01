```python
from janome.tokenizer import Tokenizer

def my_Janome(text):
  t = Tokenizer()
  results = t.tokenize(text)
  for token in results:
    print(token)
```

```python
genbun[0][1]
```

```python
my_Janome(genbun[0][1])
```

```python
def wakati(text):
  t = Tokenizer()
  results = t.tokenize(text)
  words = []
  for token in results:
    words.append(token.surface) # 表層形
  return words
```

```python
words_0 = wakati(genbun[0][1])
print(words_0)
```

```python
%%time
genbun_wakati = []
for i in range(genbun.shape[0]):
  w = wakati(genbun[i][1])
  genbun_wakati.append(w)
print(genbun_wakati)
```
