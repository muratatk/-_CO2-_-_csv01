```python
def wakati(text):
    t = Tokenizer()
    results = t.tokenize(text)
    words = []
    for token in results:
        words.append(token.surface) # 表層形
    return words
```

```python
words_0 = wakati(text[0][1])
```

```python
print(words_0)
```

```python
%%time
text_wakati = []
for i in range(text.shape[0]):
    w = wakati(text[i][1])
    text_wakati.append(w)
print(text_wakati)
```

```python
import itertools # 2次元リストを1次元化
```

```python
# 索引語の集合を作りたいので，その前に上の2次元リストを1次元化します．なお，途中（100個目）まで表示したあと，全体の個数を表示すると16394 個であることがわかります．
```

```python
words = list(itertools.chain.from_iterable(text_wakati))
print(words[0:100])
print(len(words))
```

```python
# 次のコードを実行することで単語（索引語）の集合を作ることができます．この結果，1162の索引語が使われていることがわかりました．
```

```python
vocab = sorted(set(words))
print(vocab)
print(len(vocab))
```

```python
import numpy as np
```

```python
# 索引後の頻度 TF (Term Frequency) を求める関数を定義します．
```

```python
def get_TF(text_wakati, vocab):
    """
    TFは索引語の出現頻度のこと．
    """
    n_docs = len(text_wakati)
    n_vocab = len(vocab)
 
    # 行数 = 文書数， 列数 = 登録集合数 で tf を初期化する
    tf = np.zeros((n_docs, n_vocab))
 
    for i in range(n_docs):
        for w in text_wakati[i]:
            tf[i, vocab.index(w)] += 1
    return tf
```

```python
# 上で定義した関数 get_TF に分かち書きした結果と索引語一覧を与えて，索引語頻度を求めます
```

```python
tf = get_TF(text_wakati, vocab)
print(tf)
```
