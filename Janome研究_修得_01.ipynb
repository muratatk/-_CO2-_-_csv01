```python
#ユーザー辞書をcsvファイル化する
#　https://zenn.dev/megane_otoko/articles/055_janome_user_dictionary

from google.colab import auth
from google.auth import default
import gspread
import pandas as pd


# 対象のGoogle SpreadSheetのURLを添付
ulr = "https://docs.google.com/spreadsheets/d/15aUokWsVqxr094FcGr_ttJil7phDEImCxM157vlo_Rg/edit#gid=0" # アドレスバーからそのままコピペ

# URLからSpreadSheet の固有キーを取得
key = ulr.split("/")[5] # 5番目がキーじゃないときは適宜修正してください

# 認証処理
# https://qiita.com/toraaa/items/36b4a3daa84619dd7f2e 認証処理方法が変わった

auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Google SpreadSheet からデータを取得
sh = gc.open_by_key(key)
worksheet = sh.get_worksheet(0)

# DataFrameに変換
df = pd.DataFrame(worksheet.get_all_values())

df.to_csv("/content/drive/MyDrive/Python_Content/user_dictionary.csv", header=False, index=False)
```

```python
#　一番基本のTokinize（分かち書き）+品詞確定

# https://note.com/kiyo_ai_note/n/ne1fe4f7a4f2a

from janome.tokenizer import Tokenizer
t = Tokenizer()

f = open('/content/drive/MyDrive/Python_Content/text_kadai_tikan.txt', 'r',encoding="Shift-JIS")

text = f.read()

f.close()

for token in t.tokenize(text):
    print(token)
```

```python
#　正規化する
#　Unicode正規化（ユニコードせいきか、英語: Unicode normalization）とは、等価な文字や文字の並びを統一的な内部表現に変換することでテキストの比較を容易にする、テキスト正規化処理の一種である

#  https://nanjamonja.net/archives/793

#　便利すぎるjanomeのAnalyzerモジュール【自然言語処理】

from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
from janome.charfilter import *
from janome.tokenfilter import *


f = open('/content/drive/MyDrive/Python_Content/kadai_zenbun.txt', 'r', encoding='UTF-8')

text = f.read()

f.close()

# Analyzerフレームワークの使い方 https://note.nkmk.me/python-janome-tutorial/


char_filters = [UnicodeNormalizeCharFilter(),
                RegexReplaceCharFilter('<.*?>', '')] #Unicodeを正規化することで表記ゆれを吸収, #正規表現パターンにマッチした文字列を置換

token_filters =[UpperCaseFilter()]  #アルファベット英字を大文字にする→→→全角･半角　大文字･小文字変換には別途str関数がある　https://note.nkmk.me/python-capitalize-lower-upper-title/

# t = Tokenizer()

analyzer = Analyzer(char_filters=char_filters, token_filters=token_filters )

for token_seikika_zumi in analyzer.analyze(text):
  print(token_seikika_zumi)
```
